```python
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import numpy as np
##Reading and displaying original image 
img = plt.imread('data/groupselfie.jpg')

plt.imshow(img)
```

```python
img.shape
```

```python
##converting image to grayscale by averaging 
graydouble = np.double(img)
grays = (graydouble[:,:,0]+ graydouble[:,:,1] + graydouble[:,:,2]) / 3
convert_to_uint8 = np.uint8(grays)
plt.imshow(convert_to_uint8)
```

```python
convert_to_uint8.shape
```

```python
##applying only sufficient colors to the image
im_color_map = cv2.applyColorMap(convert_to_uint8, cv2.COLORMAP_JET)
plt.imshow(im_color_map)
```

```python
##cut part of image
cut_face = img[300:700,400:800]
plt.imshow(cut_face)
```

```python
##identifying cut part in the original image
img[300:700,400:800,2] = 0
plt.imshow(img)
```

```python
##nigating image
negated_image = 255 - grays[300:700,400:800]
plt.imshow(negated_image)
```

```python
##Threshold color in image
ret, thresh = cv2.threshold(img, 150, 255, cv2.THRESH_BINARY)
plt.imshow(thresh)
```

```python
##converting thresholed image to grayscale
change_color_map_to_gray = cv2.cvtColor(thresh, cv2.COLOR_BGR2GRAY)
plt.imshow(change_color_map_to_gray)
```

```python
##factoring colors
selfie_to_grayscale= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
selfiedouble = selfie_to_grayscale.astype(np.double)
selfiefactor = selfiedouble*.25
##plt.imagesc(selfiefactor)
plt.imshow(selfiefactor)
```
